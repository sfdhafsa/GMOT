```python
import gradio as gr

def text_to_speech(text):
    # Assuming 'synthesizer' is already defined from a previous cell
    speech = synthesizer(text, forward_params={"do_sample": True})

    # Scale and convert the audio data to 16-bit integers
    audio_data = speech["audio"][0] # Get the audio data from the list
    scaled_audio = np.int16(audio_data / np.max(np.abs(audio_data)) * 32767)

    # Assuming 'scipy.io.wavfile' is imported as 'scipy.io'
    scipy.io.wavfile.write("bark_out.wav", rate=speech["sampling_rate"], data=scaled_audio)

    return "bark_out.wav"

iface = gr.Interface(fn=text_to_speech, inputs="text", outputs="audio")
```

```python
import gradio as gr
from transformers import AutoProcessor, BarkModel
import torch

# Load model
processor = AutoProcessor.from_pretrained("suno/bark-small")
model = BarkModel.from_pretrained("suno/bark-small")

def generate_speech(text):
    # Process text
    inputs = processor(text=[text], return_tensors="pt")

    # Generate audio tensor
    speech_values = model.generate(**inputs, do_sample=True)

    # Bark returns a waveform tensor with shape [1, seq_len]
    # Convert to numpy array for Gradio
    audio_array = speech_values.squeeze().cpu().numpy()

    # Return as Gradio audio (needs sample rate)
    return (16000, audio_array)  # Bark uses 16 kHz by default

# Gradio interface
iface = gr.Interface(
    fn=generate_speech,
    inputs="text",
    outputs=gr.Audio(type="numpy")
)

iface.launch(share=True)
```
